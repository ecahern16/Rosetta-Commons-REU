```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
def total_score_vs_rmsd(csv):
    csv_file = pd.read_csv(csv)
    print(csv_file)
    fig, ax = plt.subplots(1,1,figsize=(10,7.5))

    kwargs_title={'fontsize':30, 'weight':'bold', 'color':'0.2'}
    kwargs_label={'fontsize':30, 'weight':'bold', 'color':'0.2'}
    kwargs_ticks={'fontsize':20, 'weight':'bold', 'color':'0.2'}
    kwargs_plots={"color":"b","alpha":0.7,"lw":1, "edgecolors":None,"s":10}


    # Axes
    plt.title("Relationship Between Total Score and RMSD\n",**kwargs_title)
    plt.ylabel("Total Score\n",**kwargs_label)
    plt.xlabel("\nRMSD",**kwargs_label)
    plt.yticks(**kwargs_ticks)
    plt.xticks(**kwargs_ticks)
    plt.xlim(1,0)
    #ax.set_xticklabels(np.arange(5,96,5))

    for axis in ['bottom','left']:
        ax.spines[axis].set_linewidth(2.5)
        ax.spines[axis].set_color('0.2')

    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.tick_params(width=2.5,length=5,pad=10, color='0.2')
    

    # plt.plot(norm_var,label='variance',color='#9b2226',**kwargs_plots)
    # plt.plot(norm_var,label='variance',color='#ae2012',**kwargs_plots)
    # plt.plot(norm_lddt,label='lddt',color='gray',**kwargs_plots)
    plt.scatter(csv_file["rmsd"], csv_file["total_score"], color='blue',**kwargs_plots)
```

```python
def score_vs_fnat(csv, title):

    csv_file = pd.read_csv(csv)
    fig, ax = plt.subplots(1,1,figsize=(10,7.5))

    kwargs_title={'fontsize':30, 'weight':'bold', 'color':'0.2'}
    kwargs_label={'fontsize':30, 'weight':'bold', 'color':'0.2'}
    kwargs_ticks={'fontsize':20, 'weight':'bold', 'color':'0.2'}
    kwargs_plots={"alpha":0.7,"lw":1, "edgecolors":None,"s":10}


    # Axes
    plt.title("Relationship Between i_sc and Fnat:\n" + title + "\n",**kwargs_title)
    plt.ylabel("I_sc\n",**kwargs_label)
    plt.xlabel("\nFnat",**kwargs_label)
    plt.yticks(**kwargs_ticks)
    plt.xticks(**kwargs_ticks)

    for axis in ['bottom','left']:
        ax.spines[axis].set_linewidth(2.5)
        ax.spines[axis].set_color('0.2')

    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.tick_params(width=2.5,length=5,pad=10, color='0.2')

    # plt.plot(norm_var,label='variance',color='#9b2226',**kwargs_plots)
    # plt.plot(norm_var,label='variance',color='#ae2012',**kwargs_plots)
    # plt.plot(norm_lddt,label='lddt',color='gray',**kwargs_plots)
    plt.scatter(csv_file["Fnat"], csv_file["I_sc"], **kwargs_plots)
    plt.gca().invert_xaxis()
    plt.savefig('Isc_rmsd_' + title +'.png',dpi=300,bbox_inches='tight')
    
    # Legend
    # plt.rc('font',family='Arial',weight='bold')
    # plt.rc('legend' )
    # l = plt.legend(loc=2, prop={'size': 25})
    # for text in l.get_texts():
    #     text.set_color("0.2")

```
